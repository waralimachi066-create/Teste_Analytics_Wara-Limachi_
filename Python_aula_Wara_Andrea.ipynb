```python
'''Receba três valores reais (a, b e c, que serão os lados de um triângulo) e retorne a área usando a fórmula de Heron.
O valor da área calculada deve ser impresso com duas casas decimais de precisão.
Fórmula de Heron: A=√s(s−a)(s−b)(s−c)−−−−−−−−−−−−−−−−−√
, em que s=a+b+c/2'''
import numpy as np
a = float(input())
b = float(input())
c = float(input())
cal_s = ((a + b + c) / 2) #Fazendo o calculo separado
cal_a = (cal_s - a)
cal_b = (cal_s - b)
cal_c = (cal_s - c)
cal_f = cal_s * cal_a * cal_b * cal_c
heron_A = np.sqrt(cal_f) #Calculo final usando a biblioteca
print(f"{heron_A:.2f}") #Para ter apenas duas casas decimais
```

```python
'''Considere a equação de segunda grau f(x)=x2−2
Lembre que em uma equação de segundo grau f(x)=ax2+bx+c
, as raízes são:
delta=b∗b−4∗a∗c
raiz1=(−1∗b+math.sqrt(delta))/(2∗a)
raiz2=(−1∗b−math.sqrt(delta))/(2∗a)
Escreva instruções para mostrar a maior raiz desta equação, ou seja, o ponto mais à direita f(x)=0
. Imprimir a maior raiz com duas casas decimais, com print("%.2f" % raizMaior).
Como o objetivo desta atividade é apenas aprender a enviar para avaliação automática,
a solução de um problema que envolve raízes de equação de segundo grau, a resposta está disponível a seguir, junto com comentários. Basta copiar e colar
#solução'''

import numpy as np
import math

# para  função f(x) = x^2 - 2, temos que a=1, b=0 e c=-2

a = 1
b = 0
c = -2

# calculamos as duas raizes

delta = b*b - 4*a*c
raiz1 = (-1 * b + math.sqrt(delta)) / (2*a)
raiz2 = (-1 * b - math.sqrt(delta)) / (2*a)

# np.max([raiz1,raiz2]) encontra maior entre raiz1 e raiz2.  Atribuímos este valor à variável chamada "maior"

maior = np.max([raiz1,raiz2])

# impressão de valor da variável "maior"

print("%.2f " % maior)
```

```python
"""
Regras do jogo:
Você deve simular um jogo de apostas em que o jogador começa com uma quantia de dinheiro definida pelo usuário.
A cada rodada, ele faz uma aposta de valor fixo, também fornecido pelo usuário.
A simulação deve ser reprodutível, ou seja, a semente para o gerador de números aleatórios será fornecida como entrada.

O jogador começa com um valor inicial de dinheiro.
Em cada rodada, ele aposta um valor fixo.
Há 40% de chance de ganhar (e receber esse valor como lucro) e 60% de chance de perder (e perder o valor apostado).
A simulação termina após um número de rodadas definidas, ou antes, se o jogador não tiver dinheiro suficiente para apostar.

Entrada:
O programa recebe quatro números inteiros, que correspondem respectivamente à semente do gerador aleatório,
 ao valor inicial disponível, ao valor de cada aposta e ao número total de rodadas a serem jogadas.

Saída:
O programa deve imprimir uma mensagem:

"Lucro, ficando com R reais", se o jogador obtiver lucro,
"Prejuízo, ficando com X reais", se o jogador perder dinheiro ao final de todas as apostas, ou
"Saiu com o dinheiro inicial" se não tiver lucro nem prejuízo.
Entrada
1
100
10
20

Saída
Prejuízo, ficando com 40 reais
"""
semente = int(input())
valor_inicial = int(input())
valor_fixo = int(input())
rodadas = int(input())
rd.seed(semente)
valor_acumulado = valor_inicial

for rodadadojogo in range(1,rodadas + 1):
    if valor_acumulado < valor_fixo:
        break
    resultado = rd.choices(["Venceu","Perdeu"],weights=[0.4, 0.6])
    if resultado == "Venceu":
        valor_acumulado += valor_fixo
    else:
        valor_acumulado -= valor_fixo

if valor_acumulado > valor_inicial:
    print("Lucro, ficando com",valor_acumulado,"reais")
elif valor_acumulado < valor_inicial:
    print("Prejuízo, ficando com",valor_acumulado,"reais")
else:
    print("Saiu com o dinheiro inicial")
```

```python
import random as rd

#Quatro valores para a entrada
semente = int(input())
valor_inicial = int(input())
valor_fixo = int(input())
rodadas = int(input())

#Fixar a semente
rd.seed(semente)
#Para fazer a soma do valor em outra varíavel
valor_acumulado = valor_inicial
ganhar = 0.4
#Para cada rodadas em range(rodadas)
for rodada in range(rodadas):
    #Vamos colocar essa condição no começo para caso o jogados não tenha mais dinheiro para apostar o mínimo
    if valor_acumulado < valor_fixo:
        break  #Se não conseguir mais apostar
    resultado = rd.random() #Vamos usar a condição rd.random para utilizar o ganho armazenado na variável ganhar, sendo que o ramdom nos dá um valor entre 0 e 1, e o valor que queremos é 0.4 ou seja 40%
    if resultado < ganhar: #Se não cair entre 0.4 ou seja 40%, ele vai ser entre 60%, ou seja, perdeu
        valor_acumulado = valor_acumulado + valor_fixo
    else:
        valor_acumulado = valor_acumulado - valor_fixo
#A condição final para ver se o jogador perdeu o ganhou, ou manteve o mesmo valor
if valor_acumulado > valor_inicial:
    print("Lucro, ficando com", valor_acumulado, "reais")
elif valor_acumulado < valor_inicial:
    print("Prejuízo, ficando com", valor_acumulado, "reais")
else:
    print("Saiu com o dinheiro inicial")
```
