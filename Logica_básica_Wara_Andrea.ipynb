```python
A = float(input())
B = float(input())
C = float(input())
if (A >= B+ C):
    print("NAO FORMA TRIANGULO")
else: #Verifica-se que as intruções são quase as mesmas que o "TRIANGULO OBTUSANGULO", e como está resposta é isolada, vamos colocar como a principal, e por isso se não for "else"
    if (A ** 2 > B ** 2 + C ** 2):
        print("TRIANGULO OBTUSANGULO")
    if (A ** 2 < B ** 2 + C ** 2):
        print("TRIANGULO ACUTANGULO")
    if (A == B == C):
        print("TRIANGULO EQUILATERO")
    if (A ** 2 == B ** 2 + C ** 2):
        print("TRIANGULO RETANGULO")
    if (A == B and A != C) or (A == C and A != B) or (B == C and B != A): #Somente dois lados estão iguais então usei o != para o um que resta ser diferente dos outros dois
        print("TRIANGULO ISOSCELES")
```
